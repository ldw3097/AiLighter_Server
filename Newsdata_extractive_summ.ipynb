# Train

**Arguments**

-mode train (train, validate, test) # 학습 모드 설정

-encoder classifier (classifier, transformer, RNN) # 학습 방법 설정

-dropout 0.1 # dropout rate

-bert_data_path ../bert_data/korean # bert.pt파일 저장되어있는 경로(pt파일 앞의 단어까지 경로에 쓸 것)

-model_path ../models/bert_classifier_sample # train의 경우 학습된 모델 저장

-visible_gpus 0 # -1일 경우 cpu, 0이상일 경우 cuda 사용

-train_steps 학습 진행할 총 step의 수 (예: 50000)

-save_checkpoint_steps 몇 step마다 pt파일 형태의 checkpoint를 저장할지 지정 (예: 10000)

-warmup_steps 학습할때 학습률이 점점 늘어났다가 줄어드는데 그 늘어나는게 몇 step까지 적용되는지 지정

-batch_size: 한번에 몇개의 토큰(문장)을 읽을지 결정함

```python
!python train.py -mode train -encoder transformer -dropout 0.1 -bert_data_path /home/dongwon/korbertsum/bert_data/korean -model_path /home/dongwon/korbertsum/models/bert_classifier -lr 2e-3 -visible_gpus 0 -gpu_ranks 0 -world_size 1 -report_every 500 -save_checkpoint_steps 10000 -batch_size 1000 -decay_method noam -train_steps 50000 -accum_count 1 -log_file /home/dongwon/korbertsum/logs/bert_classifier -use_interval true -warmup_steps 8000 -bert_model /home/dongwon/001_bert_morp_pytorch -bert_config_path /home/dongwon/001_bert_morp_pytorch/bert_config.json -temp_dir .
```
